```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler

# Wczytanie danych z pliku CSV
dataset = pd.read_csv('/Users/pawelwrzesinski/desktop/Kodilla/titanic.csv')

# Odczyt wartości zmiennej Fare
fare_values = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values

# Statystyczna analiza zmiennej Fare
# Liczenie średniej
mean_fare = np.round(np.mean(fare_values), 2)
print(f"Średnia opłata (Fare): {mean_fare}")

# Liczenie mediany
median_fare = np.median(fare_values)
print(f"Mediana opłaty (Fare): {median_fare}")

# Liczenie mody
mode_fare = stats.mode(fare_values, keepdims=True)
print(f"Moda opłaty (Fare): {mode_fare.mode[0]}, liczba wystąpień: {mode_fare.count[0]}")

# Liczenie kwartyli
q0 = np.quantile(fare_values, 0.0)  # minimum
q1 = np.quantile(fare_values, 0.25)  # pierwszy kwartyl
q3 = np.quantile(fare_values, 0.75)  # trzeci kwartyl
q4 = np.quantile(fare_values, 1.0)  # maksimum
print(f"Q0 (Min): {q0}")
print(f"Q1 (25%): {q1}")
print(f"Q2 (Mediana): {median_fare}")
print(f"Q3 (75%): {q3}")
print(f"Q4 (Max): {q4}")

# Liczenie zakresu
range_fare = max(fare_values) - min(fare_values)
print(f"Zakres opłaty (Fare): {range_fare}")

# Liczenie rozstępu międzykwartylowego - IQR
iqr_fare = q3 - q1
print(f"Rozstęp międzykwartylowy (IQR) opłaty (Fare): {iqr_fare}")

# Liczenie wariancji
variance_fare = np.var(fare_values, ddof=1)
print(f"Wariancja opłaty (Fare): {variance_fare}")

# Liczenie odchylenia standardowego
std_deviation_fare = np.std(fare_values, ddof=1)
print(f"Odchylenie standardowe opłaty (Fare): {std_deviation_fare}")

# Wizualizacja danych
# Histogram zmiennej Fare
plt.hist(fare_values, bins=20, color='skyblue', edgecolor='black')
plt.axvline(x=mean_fare, color='red', label='Średnia')
plt.axvline(x=median_fare, color='green', label='Mediana')
plt.axvline(x=mode_fare.mode[0], color='orange', label='Moda')
plt.axvline(x=q1, color='black', linestyle='--', label='Q1')
plt.axvline(x=q3, color='black', linestyle='--', label='Q3')
plt.xlabel('Fare')
plt.ylabel('Liczba pasażerów')
plt.legend()
plt.title('Histogram opłaty (Fare)')
plt.show()

# Wykres pudełkowy (boxplot) zmiennej Fare
plt.boxplot(fare_values, vert=False, patch_artist=True)
plt.title('Boxplot opłaty (Fare)')
plt.xlabel('Fare')
plt.show()

```
