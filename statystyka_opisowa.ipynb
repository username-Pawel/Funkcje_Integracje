```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# Tworzenie danych z rozkładu normalnego
mean, std_dev = 3, 1
data_100 = np.random.normal(mean, std_dev, 100)
data_10000 = np.random.normal(mean, std_dev, 10000)

# Średnia
mean_100 = np.mean(data_100)
# Mediana
median_100 = np.median(data_100)
# Moda
mode_100 = stats.mode(data_100, keepdims=True)
# Kwartyle
q1_100 = np.percentile(data_100, 25)
q3_100 = np.percentile(data_100, 75)
# Zakres
range_100 = np.max(data_100) - np.min(data_100)
# Rozstęp międzykwartylowy
iqr_100 = q3_100 - q1_100
# Wariancja
variance_100 = np.var(data_100, ddof=1)
# Odchylenie standardowe
std_dev_100 = np.std(data_100, ddof=1)
# Asymetria rozkładu
skewness_100 = stats.skew(data_100)

# Wyświetlanie wyników
print("Statystyki opisowe dla próby 100")
print(f"Średnia: {mean_100}")
print(f"Mediana: {median_100}")
print(f"Moda: {mode_100[0][0]} (ilość wystąpień: {mode_100[1][0]})")
print(f"Kwartyl Q1: {q1_100}")
print(f"Kwartyl Q3: {q3_100}")
print(f"Zakres: {range_100}")
print(f"Rozstęp międzykwartylowy: {iqr_100}")
print(f"Wariancja: {variance_100}")
print(f"Odchylenie standardowe: {std_dev_100}")
print(f"Asymetria rozkładu: {skewness_100}")

# Średnia
mean_10000 = np.mean(data_10000)
# Mediana
median_10000 = np.median(data_10000)
# Moda
mode_10000 = stats.mode(data_10000, keepdims=True)
# Kwartyle
q1_10000 = np.percentile(data_10000, 25)
q3_10000 = np.percentile(data_10000, 75)
# Zakres
range_10000 = np.max(data_10000) - np.min(data_10000)
# Rozstęp międzykwartylowy
iqr_10000 = q3_10000 - q1_10000
# Wariancja
variance_10000 = np.var(data_10000, ddof=1)
# Odchylenie standardowe
std_dev_10000 = np.std(data_10000, ddof=1)
# Asymetria rozkładu
skewness_10000 = stats.skew(data_10000)

# Wyświetlanie wyników
print("Statystyki opisowe dla próby 10 000")
print(f"Średnia: {mean_10000}")
print(f"Mediana: {median_10000}")
print(f"Moda: {mode_10000[0][0]} (ilość wystąpień: {mode_10000[1][0]})")
print(f"Kwartyl Q1: {q1_10000}")
print(f"Kwartyl Q3: {q3_10000}")
print(f"Zakres: {range_10000}")
print(f"Rozstęp międzykwartylowy: {iqr_10000}")
print(f"Wariancja: {variance_10000}")
print(f"Odchylenie standardowe: {std_dev_10000}")
print(f"Asymetria rozkładu: {skewness_10000}")

# Histogram dla próby 100
plt.hist(data_100, bins=20, color='blue', alpha=0.7)
plt.title("Histogram dla próby 100")
plt.xlabel("Wartość")
plt.ylabel("Częstość")
plt.show()

# Histogram dla próby 10 000
plt.hist(data_10000, bins=20, color='green', alpha=0.7)
plt.title("Histogram dla próby 10 000")
plt.xlabel("Wartość")
plt.ylabel("Częstość")
plt.show()

# Wykres pudełkowy
plt.boxplot([data_100, data_10000], vert=False, labels=["Próba 100", "Próba 10 000"])
plt.title("Wykres pudełkowy dla różnych rozmiarów prób")
plt.xlabel("Wartość")
plt.show()

# Dzięki większej próbie (10 000) wartości statystyk są bardziej stabilne i lepiej odwzorowują teoretyczne cechy rozkładu normalnego,
#  co sprawia, że statystyki takie jak średnia, mediana, kwartyle, odchylenie standardowe i asymetria są bliższe teoretycznym wartościom. 
# Dla mniejszych prób (100) te statystyki mogą być bardziej zmienne, a różnice te mogą wynikać z mniejszej reprezentatywności próby i wpływu wartości odstających.
```
